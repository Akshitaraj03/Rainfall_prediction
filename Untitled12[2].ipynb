```python
#data preprocessing
#dropping highly correlated column
df=df.drop(columns=['maxtemp', 'temparature', 'mintemp'])
df.head()
```

```python
df["rainfall"].value_counts()
```

```python
#separate majority and minority class (1 & 0)
df_majority = df[df.rainfall==1]
df_minority = df[df.rainfall==0]
```

```python
print(df_majority.shape)
print(df_minority.shape)
```

```python
#downsample majority class to match minority count
df_majority_downsampled = resample(df_majority, replace=False, n_samples=len(df_minority), random_state=42)
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled=pd.concat([df_majority_downsampled, df_minority])
```

```python
df_downsampled.shape
```

```python
df_downsampled.head()
```

```python
#lets suffle the final dataset
df_downsampled=df_downsampled.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
x=df.drop(columns=["rainfall"])
y=df["rainfall"]
print(x)
print(y)
```

```python
#split the data into training data and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#model training
rf=RandomForestClassifier(random_state=42)
```

```python

print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
x_test.head()
```

```python
from sklearn.preprocessing import StandardScaler,Normalizer #data standarization
from sklearn.impute import SimpleImputer
sc=StandardScaler()
nr=Normalizer()
imputer = SimpleImputer(strategy='mean')
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_train = imputer.fit_transform(x_train) # Impute missing values in x_train
x_test = imputer.transform(x_test)
x_train=nr.fit_transform(x_train)
x_test=nr.transform(x_test)
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```
